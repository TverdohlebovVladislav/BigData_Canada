```python
import numpy as np
import pandas as pd
from pandas.io.excel import ExcelWriter
import os
```

Ниже код для генерации "каркаса" таблиц согласно ER-диаграмме (генерируются только поля PK и FK) и заносятся в таблицу excel. Чтобы перезаписать данные в файл, убедитесь что его нет по указанному пути в переменной path (временно, чуть позже подправлю). 

```python
# Path to save EXCEL
path = "file.xlsx"

# Minimum count of data in every table
max_count_customer: int = 100
max_count_product: int = 30
max_count_product_inst: int = 100
max_count_costed_event: int = 10485
max_count_costed_charge: int= 1233
max_count_costed_payment: int = 12442

# ---
# 1. Create dataframe - Customer
customer_id_PK: list = [i for i in range(1, max_count_customer)]
CustomerDf: list = pd.DataFrame(
    {
        "customer_id": pd.Series(customer_id_PK, name="customer_id", dtype="int"),
        # Add other columns here...
    }
)
CustomerDf.set_index('customer_id', inplace=True)


# ---
# 2. Create dataframe - Product
product_id: list  = [i for i in range(1, max_count_product)] 
ProductDf = pd.DataFrame(
    {
        "product_id": pd.Series(product_id, name="product_id", dtype="int"),
        # Add other columns here...
    }
)
ProductDf.set_index('product_id', inplace=True)


# ---
# 3. Create dataframe - Product_instance
product_instance_id_PK: list  = [i for i in range(1, max_count_product_inst)]
customer_id_FK: list  = np.random.randint(1, max_count_customer, size=max_count_product_inst - 1)
product_id_FK: list  = np.random.randint(1, max_count_product, size=max_count_product_inst - 1) 
Product_instanceDf = pd.DataFrame(
    {
        "product_instance_id_PK": pd.Series(product_instance_id_PK, name="product_instance_id_PK", dtype="int"),
        "customer_id_FK": pd.Series(customer_id_FK, name="customer_id_FK", dtype="int"),
        "product_id_FK": pd.Series(product_id_FK, name="product_id_FK", dtype="int"),
        # Add other columns here...
    }
)
Product_instanceDf.set_index('product_instance_id_PK', inplace=True)


# ---
# 4. Create dataframe - Costed_event
event_id_PK: list  = [i for i in range(1, max_count_costed_event)] 
product_instance_id_FK_from_ce: list  = np.random.randint(1, max_count_product_inst, size=max_count_costed_event - 1)
Costed_eventDf = pd.DataFrame(
    {
        "event_id_PK": pd.Series(event_id_PK, name="event_id_PK", dtype="int"),
        "product_instance_id_FK": pd.Series(product_instance_id_FK_from_ce, name="product_instance_id_FK", dtype="int"),
        # Add other columns here...
    }
)
Costed_eventDf.set_index('event_id_PK', inplace=True)


# ---
# 5. Create dataframe - Charge
charge_id_PK: list  = [i for i in range(1, max_count_costed_charge)] 
product_instance_id_FK_charge: list   = np.random.randint(1, max_count_product_inst, size=max_count_costed_charge - 1)
ChargeDf = pd.DataFrame(
    {
        "charge_id_PK": pd.Series(charge_id_PK, name="charge_id_PK", dtype="int"),
        "product_instance_id_FK": pd.Series(product_instance_id_FK_charge, name="product_instance_id_FK", dtype="int"),
        # Add other columns here...
    }
)
ChargeDf.set_index('charge_id_PK', inplace=True)


# ---
# 6. Create dataframe - Payment
payment_id_PK: list  = [i for i in range(1, max_count_costed_payment)]
customer_id_FK_in_payment: list = np.random.randint(1, max_count_customer, size=max_count_costed_payment - 1)
PaymentDf = pd.DataFrame(
    {
        "payment_id_PK": pd.Series(payment_id_PK, name="payment_id_PK", dtype="int"),
        "customer_id_FK": pd.Series(customer_id_FK_in_payment, name="customer_id_FK", dtype="int"),
        # Add other columns here...
    }
)
PaymentDf.set_index('payment_id_PK', inplace=True)


# ---
# Save to EXCEL
with ExcelWriter(path, mode="a" if os.path.exists(path) else "w") as writer:
    CustomerDf.to_excel(writer, sheet_name="Customer")
    ProductDf.to_excel(writer, sheet_name="Product")
    Product_instanceDf.to_excel(writer, sheet_name="Product_instanc")
    Costed_eventDf.to_excel(writer, sheet_name="Costed_event")
    ChargeDf.to_excel(writer, sheet_name="Charge")
    PaymentDf.to_excel(writer, sheet_name="Payment")
```
